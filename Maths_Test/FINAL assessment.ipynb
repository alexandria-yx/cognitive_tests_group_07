```python
from IPython.display import display, Image, clear_output, HTML
import time
import ipywidgets as widgets
from jupyter_ui_poll import ui_events
import requests
from bs4 import BeautifulSoup
import json 


data_consent_info = """DATA CONSENT INFORMATION:

Please read:

We wish to record your response data
to an anonymised public data repository. 
Your data will be used for educational teaching purposes
practising data analysis and visualisation.

Please type   yes   in the box below if you consent to the upload."""

print(data_consent_info)
result = input("> ") 

if result == "yes": 
    print("Thanks for your participation.")
    print("Please contact a.fedorec@ucl.ac.uk")
    print("If you have any questions or concerns")
    print("regarding the stored results.")
    
    time.sleep(6)
    clear_output(wait=False)
    
else: 
    # end code execution by raising an exception
    raise(Exception("User did not consent to continue test."))

id_instructions = """

Enter your anonymised ID

To generate an anonymous 4-letter unique user identifier please enter:
- two letters based on the initials (first and last name) of a childhood friend
- two letters based on the initials (first and last name) of a favourite actor / actress

e.g. if your friend was called Charlie Brown and film star was Tom Cruise
     then your unique identifier would be CBTC
"""

print(id_instructions)
user_id = input("> ")

print("User entered id:", user_id)

def send_to_google_form(data_dict, form_url):
    ''' Helper function to upload information to a corresponding google form 
        You are not expected to follow the code within this function!
    '''
    form_id = form_url[34:90]
    view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
    post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'

    page = requests.get(view_form_url)
    content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
    content = content.text[27:-1]
    result = json.loads(content)[1][1]
    form_dict = {}
    
    loaded_all = True
    for item in result:
        if item[1] not in data_dict:
            print(f"Form item {item[1]} not found. Data not uploaded.")
            loaded_all = False
            return False
        form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
    post_result = requests.post(post_form_url, data=form_dict)
    return post_result.ok

event_info = {
    'description': '',
}

def wait_for_event(timeout=-1, interval=0.001, max_rate=20, allow_interupt=True):    
    start_wait = time.time()

    event_info['description'] = ""
    
    n_proc = int(max_rate*interval)+1
    
    with ui_events() as ui_poll:
        keep_looping = True
        while keep_looping==True:
            ui_poll(n_proc)

            if (timeout != -1) and (time.time() > start_wait + timeout):
                keep_looping = False
          
            if allow_interupt==True and event_info['description']!="":
                keep_looping = False
                
            time.sleep(interval)
    
    return event_info
    
def register_event(btn):
    event_info['description'] = btn.description
    return


 

color = "blue"
color1 = "black"
score = 0
message = "Welcome to the maths ability test!"

style = f"color: {color}; font-size: 50px; font-family: verdana;"
html_out = HTML(f"""<span style='{style}'>{message}</span>""")

display(html_out)
time.sleep(3)

message1 = "<br/>You will be asked to perform a series of numerical questions. The test will increase in difficulty with each question. Complete the calculations in the order as the numbers show up, not according to the order of operations. The test will last approximately 3 minutes. You will be given your score at the end, and an average will be taken of your time taken to answer each question. Good luck!"

style = f"color: {color}; font-size: 20px; font-family: verdana;"
html_out1 = HTML(f"""<span style='{style}'>{message1}</span>""")

display(html_out1)
time.sleep(8)
clear_output(wait=False)

score = 0
message2 = "<br/>Let's start with a practice question."

style = f"color: {color}; font-size: 20px; font-family: verdana;"
html_out2 = HTML(f"""<span style='{style}'>{message2}</span>""")

display(html_out2)
time.sleep(2)

message3 = "<br/>2 + 4 =" 

style = f"color: {color1}; font-size: 20px; font-family: verdana;"
html_out3 = HTML(f"""<span style='{style}'>{message3}</span>""")

start_time = time.time()

display(html_out3)

answer = input("\nEnter the answer: ")
if int(answer) == 6:
    score += 1
    
end_time = time.time()

time_taken = end_time - start_time

print(f"You took {time_taken:.2f} seconds to answer this question.")
print(f"Your score is {score}/1.")

time.sleep(5)
clear_output(wait=False)

message4 = "<br/>Now that you have had some practice, let's begin the test." 

style = f"color: {color}; font-size: 20px; font-family: verdana;"
html_out4 = HTML(f"""<span style='{style}'>{message4}</span>""")

display(html_out4)
time.sleep(3)
clear_output(wait=False)

questions = [['2', '+ 8', '- 7', 'x 2'], ['4', '+ 16', '- 2', '÷ 2'], ['16', '÷ 4', '+ 11', '÷ 3'], ['5', 'x 3', '+ 9', '÷ 2'], ['3', 'x 7', '+ 3', '÷ 3'], ['7', 'x 6', '- 7', '+ 58'], ['4', 'x 12', '- 27', '÷ 3'],
             ['35', '÷ 7', '+ 7', 'x 8'], ['17', 'x 2', '+ 54', '÷ 2'], ['23', 'x 4', '÷ 2', '+ 4'], ['72', '÷ 8', 'x 5', '+ 69'], ['67', 'x 3', '- 99', '÷ 2'], ['89', '+ 55', '÷ 12', 'x 7'],
             ['53', 'x 3', '- 90', '+ 41']]
answers = ['6', '9', '5', '12', '8', '93', '7', '96', '44', '50', '114', '51', '84', '110']

final_score = 0 
average_time = 0
def maths_test():
    global final_score
    global average_time
    total_time_taken = 0
    for question, answer in zip(questions, answers):
        for number in question:
            print(number)
            time.sleep(2.5)
            clear_output(wait=False)
        start_time = time.time()
        ans = input("Answer?")
        if ans == answer:
            final_score += 1
        clear_output(wait=False)
        end_time = time.time()
        time_taken = end_time - start_time
        total_time_taken += time_taken
    result = f"Your final score for this test is {final_score}/14."
    average_time = total_time_taken/14
    result_time = f"You took an average of {average_time} seconds to complete each question."
    
    print(result)
    print(result_time)
    return

maths_test()

time.sleep(6)
clear_output(wait=False)

btn1 = widgets.Button(description="Hard")
btn2 = widgets.Button(description="Medium")
btn3 = widgets.Button(description="Easy")
    

btn1.on_click(register_event) 
btn2.on_click(register_event) 
btn3.on_click(register_event) 


myhtml1 = HTML("<h1>How did you find the test?</h1>")
display(myhtml1)
myhtml2 = HTML("<h2>You have 5 seconds to answer</h2>")
display(myhtml2)

panel = widgets.HBox([btn1, btn2, btn3])
display(panel)

result = wait_for_event(timeout=10)
clear_output()

if result['description']=="Hard":
    print(f"That's unfortunate. Better luck next time!")
elif result['description']=="Medium":
    print("Good effort.")
elif result['description']=="Easy":
    print("Well done you maths whizz!") 

ans1 = user_id
ans2 = final_score
ans3 = average_time
ans4 = result['description'] 

data_dict = {'user_id':ans1, 'user_score':ans2, 'user_time':ans3, 'user_eval':ans4} 

form_url = "https://docs.google.com/forms/d/e/1FAIpQLScrTdWaNfg6nzIXBBVSVOY1nSe8b2ZL0HyJ1rVZVZRElwU21Q/viewform?usp=sf_link"
send_to_google_form(data_dict, form_url)
```
