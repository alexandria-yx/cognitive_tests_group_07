```python
import time
from time import sleep
import random
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from jupyter_ui_poll import ui_events
from IPython.display import display, Image, HTML, clear_output
import requests
from bs4 import BeautifulSoup
import json
```

```python
##Assigning names to grid images and global variables

test_grid_easy = Image("EASY memory test grid.png", width = 500)
test_grid_med = Image("MEDIUM memory test grid.png", width = 500)
test_grid_hard = Image("HARD memory test grid.png", width = 500)
shape_key = Image("shape key.png", width = 400)

total_score = 0

def display_grid(grid_array):
    '''Function which displays grid array if needed'''
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.imshow(grid_array)
    plt.show()
    return



##DATA CONSENT
def data_consent():
    data_consent_info = """DATA CONSENT INFORMATION: 
    
    Please read: 
    
    We wish to record your response data 
    to an anonymised public data repository.  
    Your data will be used for educational teaching purposes 
    practising data analysis and visualisation. 
    Please type 'yes' in the box below if you consent to the upload.
    """ 
    
    print(data_consent_info) 
    
    result = input("> ")
    clear_output(wait=False)
    if result == "yes":  
        print("Thanks for your participation.") 
        print("Please contact a.fedorec@ucl.ac.uk") 
        print("If you have any questions or concerns") 
        print("regarding the stored results.") 
        time.sleep(5)
        clear_output(wait=False)
    else:  
        # end code execution by raising an exception 
        raise Exception("User did not consent to continue test.")

#GENDER INFO
def ask_gender():
    question = HTML(f"<span style='color: green; font-weight: 500; font-size: 25px'>{'Please type in your gender: Male/Female/Other'}</span>")
    display(question)
    gender = input().lower()
    clear_output(wait=False)
    return


##INSTRUCTIONS AND USER ID

instructions = ['> Welcome!', '> This is a memory test with 3 rounds.',
                '> The rounds will be in order of easy, medium, and hard difficulty.',
                '> In each round, you will have 20 seconds to memorise a grid.',
                '> Each grid contains shapes of different colours.',
                '> Pay attention to ALL the shapes, colours, and positions in the grid.',
                '> After 20 seconds you will be asked questions on the grid you just memorised.',
                '> Some grids show big shapes with small shapes inside them; unless referencing the small shapes explicitly, the questions refer to the big shapes.',
                '> All the answers are ONE WORD (e.g. blue instead of dark blue)',
                '> Good luck!']

style = 'color: green; font-weight: 500; font-size: 15px;'

def display_instructions():
    for sentence in instructions:
        display(HTML(f"<span style='{style}'>{sentence}</span>"))
        if len(sentence)<40:
            time.sleep(1.5)
        else:
            time.sleep(3.5)
    clear_output(wait=False)
    display(HTML(f"<span style='{style}'>{'Get ready for your test!'}</span"))
    time.sleep(5)
    clear_output(wait=False)
    return

user_id = ''
def ask_id():
    id_instructions = """ First, please enter your anonymised ID 
    
    To generate an anonymous 4-letter unique user identifier please enter: 
    
    - two letters based on the initials (first and last name) of a childhood friend 
    - two letters based on the initials (first and last name) of a favourite actor / actress 
    
    e.g. if your friend was called Charlie Brown and film star was Tom Cruise 
         then your unique identifier would be CBTC 
    """
    global user_id
    print(id_instructions)
    user_id = input("Enter ID here:").upper()
    clear_output(wait=False)
    print('User entered ID:', user_id)
    time.sleep(1.5)
    clear_output(wait=False)


##EASY SECTION

#Displaying easy grid

def display_easy_grid():
    display(HTML(f"<span style='{style}'>{'First up is the easy grid.'}</span>"))
    time.sleep(1.5)
    clear_output(wait=True)
    display(test_grid_easy)
    time.sleep(20)
    clear_output(wait=False)
    display(HTML(f"<span style='{style}'>{"Get ready for your questions!"}</span>"))
    time.sleep(1.5)
    clear_output(wait=False)
    return()


##Easy questions test
##Dictionary of easy questions
easy_questions = {'What colour was the triangle?':'Yellow',
                  'What shape was in the top-right grid?':'Hexagon',
                  'What shape was in red?':'Circle',
                  'What colour was the hexagon?':'Blue',
                  'What was the shape in the bottom-middle grid?':'Diamond',
                  'What colour was the arrow?':'Green',
                  'What shape was above the heart?':'Triangle',
                  'What colour was the shape between the triangle and hexagon?':'Arrow',
                 }

##Function for easy test
easy_score = 0
avg_easy_time = 0

def easy_test():
    '''this asks the easy round questions, and displays a numpy grid array if a
        position question is asked to clarify the relevant position visually. Time
        taken to answer is recorded'''
    global easy_score
    total_time = 0
    for question, answer in easy_questions.items():
        display(shape_key)
        grid_array_easy = np.zeros(shape=(2, 3), dtype= 'int64')
        if 'top-right' in question:
            grid_array_easy[0,2] = 1
            display_grid(grid_array_easy)
        if 'bottom-middle' in question:
            grid_array_easy[1,1] = 1
            display_grid(grid_array_easy)
        display(HTML(f"<span style='{style}'>{question}</span>"))
        start_time = time.time()
        user_answer = input()
        if user_answer == answer.lower():
            easy_score += 1
            clear_output(wait=False)
            display(HTML(f"<span style='{style}'>{'Correct!'}</span>"))
        else:
            clear_output(wait=False)
            display(HTML(f"<span style='{style}'>{'Incorrect!'}</span>"))
        end_time = time.time()
        question_time = end_time - start_time
        total_time += question_time
        time.sleep(1.5)
        clear_output(wait=False)
    global total_score
    global avg_easy_time
    total_score += easy_score
    avg_easy_time = total_time / len(easy_questions)
    result = f'''
    Your score for this round is {easy_score}/{len(easy_questions)} questions. 
    Your average answering time for this round is {avg_easy_time:.2f}s.'''
    display(HTML(f"<span style='{style}'>{result}</span>"))
    time.sleep(5)
    clear_output(wait=False)
    return



##MEDIUM SECTION

#Displaying medium grid
def display_med_grid():
    display(HTML(f"<span style='{style}'>{"Next is the medium grid."}</span>"))
    time.sleep(1.5)
    clear_output(wait=True)
    display(test_grid_med)
    time.sleep(20)
    clear_output(wait=False)
    display(HTML(f"<span style='{style}'>{"Get ready for your questions!"}</span>"))
    time.sleep(1.5)
    clear_output(wait=False)
    return()

##Medium questions test
##Dictionary of hard questions
med_questions = {'What shape was green?':'Cross',
                  'What colour was the heart?':'Brown',
                  'What shape was on the very left?':'Pentagon',
                  'What small shape was inside the pentagon?':'Triangle',
                  'Was there a purple shape in the grid?':'No',
                  'What shape was blue?':'Star',
                  'What colour was the small shape inside the cross?':'Red',
                  'What was the small shape inside the star?':'Diamond',
                }

##Function for medium test
med_score = 0
avg_med_time = 0

def med_test():
    global med_score
    total_time = 0
    for question, answer in med_questions.items():
        display(shape_key)
        display(HTML(f"<span style='{style}'>{question}</span>"))
        start_time = time.time()
        user_answer = input().lower()
        if user_answer == answer.lower():
            med_score += 1
            clear_output(wait=False)
            display(HTML(f"<span style='{style}'>{'Correct!'}</span>"))
        else:
            clear_output(wait=False)
            display(HTML(f"<span style='{style}'>{'Incorrect!'}</span>"))
        end_time = time.time()
        time.sleep(1.5)
        question_time = end_time - start_time
        total_time += question_time
        clear_output(wait=False)
    global total_score
    global avg_med_time
    total_score += med_score
    avg_med_time = total_time / len(med_questions)
    result = f'''
    Your score for this round is {med_score}/{len(med_questions)} questions.
    Your average answering time for this round is {avg_med_time:.2f}s.'''
    display(HTML(f"<span style='{style}'>{result}</span>"))
    time.sleep(5)
    clear_output(wait=False)
    return





##HARD SECTION

#Displaying hard grid
def display_hard_grid():
    display(HTML(f"<span style='{style}'>{"Finally, here is the hard grid."}</span>"))
    time.sleep(1.5)
    clear_output(wait=True)
    display(test_grid_hard)
    time.sleep(20)
    clear_output(wait=False)
    display(HTML(f"<span style='{style}'>{"Get ready for your questions!"}</span>"))
    time.sleep(1.5)
    clear_output(wait=False)
    return()


##Hard questions test
##Dictionary of hard questions
hard_questions = {'What colour was the circle?':'Grey',
                  'What shape was in the top-centre grid?':'Square',
                  'What shape was above the star?':'Pentagon',
                  'What shape was inside the diamond?':'Heart',
                  'What colour was the shape in the middle-left grid?':'Red',
                  'What shape was pink?':'Cross',
                  'What shape was between the pentagon and triangle?':'Circle',
                  'Was there a rectangle in the grid?':'No',
                  'What shape was in the bottom-right grid?':'Triangle',
                  'What colour was the diamond?':'Green',
                 }

##Function for hard test
hard_score = 0
avg_hard_time = 0

def hard_test():
    global hard_score
    total_time = 0
    for question, answer in hard_questions.items():
        display(shape_key)
        grid_array_hard = np.zeros(shape=(3, 3), dtype= 'int64')
        if 'top-centre' in question:
            grid_array_hard[0,1] = 1
            display_grid(grid_array_hard)
        if 'middle-left' in question:
            grid_array_hard[1,0] = 1
            display_grid(grid_array_hard)
        if 'bottom-right' in question:
            grid_array_hard[2,2] = 1
            display_grid(grid_array_hard)
        display(HTML(f"<span style='{style}'>{question}</span>"))
        start_time = time.time()
        user_answer = input().lower()
        if user_answer == answer.lower():
            hard_score += 1
            clear_output(wait=False)
            display(HTML(f"<span style='{style}'>{'Correct!'}</span>"))
        else:
            clear_output(wait=False)
            display(HTML(f"<span style='{style}'>{'Incorrect!'}</span>"))
        end_time = time.time()
        time.sleep(1.5)
        question_time = end_time - start_time
        total_time += question_time
        clear_output(wait=False)
    global total_score
    global avg_hard_time
    total_score += hard_score
    avg_hard_time = total_time / len(hard_questions)
    result = f'''
    Your score for this round is {hard_score}/{len(hard_questions)} questions.
    Your average answering time for this round is {avg_hard_time:.2f}s.'''
    display(HTML(f"<span style='{style}'>{result}</span>"))
    time.sleep(5)
    clear_output(wait=False)
    return





#####BUTTON CODE

event_info = {
    'type': '',
    'description': '',
    'time': -1
}

user_input = ''

def wait_for_event(timeout=-1, interval=0.001, max_rate=20, allow_interupt=True):    
    start_wait = time.time()

    # set event info to be empty
    # as this is dict we can change entries
    # directly without using
    # the global keyword
    event_info['type'] = ""
    event_info['description'] = ""
    event_info['time'] = -1

    n_proc = int(max_rate*interval)+1
    
    with ui_events() as ui_poll:
        keep_looping = True
        while keep_looping==True:
            # process UI events
            ui_poll(n_proc)

            # end loop if we have waited more than the timeout period
            if (timeout != -1) and (time.time() > start_wait + timeout):
                keep_looping = False
                
            # end loop if event has occured
            if allow_interupt==True and event_info['description']!="":
                keep_looping = False
                
            # add pause before looping
            # to check events again
            time.sleep(interval)
    
    # return event description after wait ends
    # will be set to empty string '' if no event occured
    return event_info

# this function lets buttons 
# register events when clicked
def register_event(btn):
    # display button description in output area
    event_info['type'] = "click"
    event_info['description'] = btn.description
    event_info['time'] = time.time()
    return

def user_opinion():
    btn1 = widgets.Button(description="Amazing")
    btn2 = widgets.Button(description="Okay")
    btn3 = widgets.Button(description="Bad")
    
    # we need to set up each button
    # to call the register_event
    # function when clicked
    btn1.on_click(register_event) 
    btn2.on_click(register_event) 
    btn3.on_click(register_event) 

    myhtml1 = HTML(f"<span style='color: green; font-weight: 500; font-size: 30px'>{'How do you think you did on this memory test?'}</span>")
    display(myhtml1)

    panel = widgets.HBox([btn1, btn2, btn3])
    display(panel)

    result = wait_for_event(timeout=60)
    clear_output()

    if result['description']!="":
        display(HTML(f"<span style='{style}'>{f"You thought your memory was {result['description'].lower()}!"}</span>"))
        global user_input
        user_input = result['description']
    else:
        print("User did not click in time")
    time.sleep(5)
    clear_output(wait=False)
    return


##GOOGLE FORMS UPLOAD CODE!!!

def send_to_google_form(data_dict, form_url):
    ''' Helper function to upload information to a corresponding google form 
        You are not expected to follow the code within this function!
    '''
    form_id = form_url[34:90]
    view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
    post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'

    page = requests.get(view_form_url)
    content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
    content = content.text[27:-1]
    result = json.loads(content)[1][1]
    form_dict = {}
    
    loaded_all = True
    for item in result:
        if item[1] not in data_dict:
            print(f"Form item {item[1]} not found. Data not uploaded.")
            loaded_all = False
            return False
        form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
    post_result = requests.post(post_form_url, data=form_dict)
    return post_result.ok
```

```python
data_consent()

ask_id()
ask_gender()
display_instructions()

display_easy_grid()
easy_test()
display_med_grid()
med_test()
display_hard_grid()
hard_test()

user_opinion()
display(HTML(f"<span style='{style}'>{f'You have now completed the memory test, and your total score was {total_score}/26 \n({((total_score/26)*100):.1f}%).'}</span>"))


##Send results to Google form and sheets
data_dict = {
    "User ID": user_id,
    "Gender": gender,
    "Easy score": f'{(easy_score/8):.2%}',
    "Average easy time": f'{avg_easy_time:.2f}',
    "Medium score": f'{(med_score/8):.2%}',
    "Average medium time": f'{avg_med_time:.2f}',
    "Hard score": f'{(hard_score/10):.2%}',
    "Average hard time": f'{avg_hard_time:.2f}',
    "Performance reflection" : user_input
}

form_url = 'https://docs.google.com/forms/d/e/1FAIpQLSfWo9DCaZPvksV-sFp0zV3piPZYZZrM5wMNhCpvUEeX2okR4w/viewform'

send_to_google_form(data_dict, form_url);
```
